```python
def display_head(df: pyspark.sql.DataFrame, n: int):
    return df.limit(n).toPandas().head(n)
```

## Zadanie 1

```python
movies_df = spark.read.csv('../data/movies.csv', header=True, inferSchema=True)

display_head(movies_df, 50)
```

```python
df = movies_df.na.fill({'budget': '$ 0', 'writer': 'Unknown', 'reviews_from_critics': 0.0})

display_head(df, 50)
```

```python
from pyspark.sql.functions import explode, split

titles_with_genres = movies_df \
    .withColumn('genre', split('genre', ', ')) \
    .select('title', explode('genre').alias('genre'))

display_head(titles_with_genres, 20)
```

```python
df = movies_df.drop('description', 'metascore', 'usa_gross_income')

display_head(df, 10)
```

```python
df = movies_df.dropna()

display_head(df, 20)
```

```python
from pyspark.sql.functions import regexp_replace
df = movies_df.withColumn('genre', regexp_replace('genre', r'Drama', r'Dramat'))

display_head(df, 20)
```

```python
from pyspark.sql.functions import regexp_extract

df = movies_df.withColumn('budget2', regexp_extract("budget", r"\S+ (\d+)", 1))

display_head(df, 50)
```

```python
from pyspark.sql.functions import coalesce

df = movies_df.withColumn('budget', coalesce('budget', lit('$ 0')))

display_head(df, 20)
```

```python
from pyspark.sql.functions import concat
df = movies_df.withColumn('title_evolution', when(col('title') == col('original_title'), None).otherwise(concat(col('original_title'), lit(' ----> '), col('title'))))

display_head(df, 50)
```

```python
df = movies_df.replace(['Italian', 'German'], ['Italiano', 'Deutsch'], 'language')

display_head(df, 20)
```

```python
from pyspark.sql.functions import corr, count_distinct, collect_set

df = movies_df.groupBy('country').agg(
    count_distinct('title').alias('films_count'),
    corr('reviews_from_users', 'reviews_from_critics').alias('reviews_numbers_corr'),
    collect_set('language').alias('languages')
).filter(col('films_count') > 5)

display_head(df, 20)
```

## Zadanie 2

```python
def prod(*values):
    return float(np.nanprod(values))

def cumsum_average(*values):
    return int(np.cumsum(values).mean())

def join_with_commas(*values):
    return str(','.join(values))
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType, StringType, IntegerType

prod_udf=udf(prod, DoubleType())
cumsum_average_udf=udf(cumsum_average, IntegerType())
join_with_commas_udf=udf(join_with_commas, StringType())

df = movies_df.dropna() \
    .withColumn('weird_double', prod_udf('reviews_from_critics', 'reviews_from_users')) \
    .withColumn('weird_int', cumsum_average_udf('duration', 'votes'))\
    .withColumn('joined_str', join_with_commas_udf('title', 'original_title', 'director'))

display_head(df, 20)
```
